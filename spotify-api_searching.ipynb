```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pprint
```

```python
class Spotify_audio_features:
    def __init__(self):
        cid = '3ff7c7512f6c4025adc2e568a7712d0a'
        secret = 'e80e183fd12a48edb4cae95b3e4cbf4a'
    
        client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
        self.sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
    
    def get_features(self, song):
        track_info = self.sp.search(q = song, type='track')
        track_id = track_info["tracks"]["items"][0]["id"]
        track_name = track_info["tracks"]["items"][0]["name"]
        artists = track_info['tracks']['items'][0]['artists'][0]['name']
        popularity = track_info['tracks']['items'][0]['popularity']
        features = self.sp.audio_features(tracks=[track_id])
        acousticness = features[0]["acousticness"]
        danceability = features[0]["danceability"]
        energy = features[0]["energy"]
        liveness = features[0]["liveness"]
        loudness = features[0]["loudness"]
        valence = features[0]["valence"]
        mode = features[0]["mode"]
        key = features[0]['key']
        instrumentalness = features[0]['instrumentalness']
        tempo = features[0]['tempo']
        time_signature = features[0]['time_signature']

        result = {"Title" : track_name,
                    "Artists" : artists,
                    "Popularity" : popularity,
                    "Acousticness" : acousticness,
                    "Danceability" : danceability,
                    "Energy" : energy,
                    "Liveness" : liveness,
                    "Loudness" : loudness,
                    "Valence" : valence,
                    "Mode" : mode,
                    "Key" : key,
                    "Instrumentalness" : instrumentalness,
                    "Tempo" : tempo,
                    "Track id" : track_id,}
        return result
```
